```python
import os
from dotenv import load_dotenv
from langchain_together import ChatTogether

# 🔸 Modern Core Imports
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.tools import Tool

# 🔐 Load API keys from .env
load_dotenv(dotenv_path=".env")
together_api_key = os.getenv("TOGETHER_API_KEY")

# 🔸 Step 1: Initialize the LLM
llm = ChatTogether(
    model="meta-llama/Meta-Llama-3-8B-Instruct-Lite", 
    temperature=0,
    together_api_key=together_api_key
)

# 🔸 Step 2: Create a prompt template
prompt = PromptTemplate.from_template("Answer clearly: {question}")

# 🔸 Step 3: Build the Chain using LCEL
# We pipe the prompt -> llm -> string output parser
qa_chain = prompt | llm | StrOutputParser()

# 🔸 Step 4: Wrap LCEL Chain as a Tool
qa_tool = Tool(
    name="Simple QA",
    # We use a lambda to pass the string query into the dict format the prompt expects
    func=lambda q: qa_chain.invoke({"question": q}),
    description="A basic LLM chain that answers clear questions"
)

# 🔸 Step 5: Use the Tool directly using .invoke()
query = "What is LangGraph in LangChain?"
answer = qa_tool.invoke(query)

# 🖨️ Output
print("\nUser Question:", query)
print("\nTool Answer:", answer)
```
