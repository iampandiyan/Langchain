```python
import os
from dotenv import load_dotenv
from langchain_together import ChatTogether

# 🔸 Modern Core Imports
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.tools import Tool

# 🔸 Fixed Tavily Import
from langchain_tavily import TavilySearch

# 🔐 Load API keys
load_dotenv(dotenv_path=".env")
together_api_key = os.getenv("TOGETHER_API_KEY")
# Note: Ensure you have TAVILY_API_KEY in your .env file!
tavily_api_key = os.getenv("TAVILY_API_KEY")

# 🔸 Initialize the LLM
llm = ChatTogether(
    model="meta-llama/Meta-Llama-3-8B-Instruct-Lite", 
    temperature=0,
    together_api_key=together_api_key
)

# ✅ Tool 1: Simple QA 
qa_prompt = PromptTemplate.from_template("Answer clearly: {question}")
qa_chain = qa_prompt | llm | StrOutputParser()

qa_tool = Tool(
    name="Simple QA",
    func=lambda q: qa_chain.invoke({"question": q}),
    description="Answer factual questions clearly"
)

# ✅ Tool 2: Summarizer 
summary_prompt = PromptTemplate.from_template("Summarize the following text:\n\n{text}")
summary_chain = summary_prompt | llm | StrOutputParser()

summary_tool = Tool(
    name="Summarizer",
    func=lambda t: summary_chain.invoke({"text": t}),
    description="Summarizes input text"
)

# ✅ Tool 3: Web Search (Fixed class name)
search_tool = TavilySearch(max_results=3)

# 🧪 Run each tool manually using .invoke()
qa_query = "What is LangGraph in LangChain?"
summary_text = """
LangGraph is a framework for building stateful multi-step agents using LangChain. 
It uses graph-based design to model agent workflows and memory.
"""
search_query = "Latest updates on GPT-4o by OpenAI"

print("\n🧠 Simple QA Tool Output:\n", qa_tool.invoke(qa_query))
print("\n📝 Summarizer Tool Output:\n", summary_tool.invoke(summary_text))
print("\n🌐 Web Search Tool Output:\n", search_tool.invoke({"query": search_query}))
```
