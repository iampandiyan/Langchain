```python
import os
from dotenv import load_dotenv
from langchain_together import ChatTogether
from langchain_core.prompts import PromptTemplate

# 🔐 Load API key
load_dotenv(dotenv_path=".env")
together_api_key = os.getenv("TOGETHER_API_KEY")

# 🔸 Initialize LLM
llm = ChatTogether(
    model="meta-llama/Meta-Llama-3-8B-Instruct-Lite", 
    temperature=0,
    together_api_key=together_api_key
)

# 🔸 Create Prompt Template
prompt = PromptTemplate.from_template("Answer clearly: {question}")

# 🔸 Create Chain using LCEL (Piping prompt into LLM)
qa_chain = prompt | llm

# 🚀 Run a query using invoke()
query = "What is LangGraph in LangChain?"
response = qa_chain.invoke({"question": query})

# 🖨️ Output
print("\nUser Question:", query)
print("\nLLM Answer:", response.content)
```
